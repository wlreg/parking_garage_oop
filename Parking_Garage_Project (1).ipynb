# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
# Your Code will go here

class ParkingGarage():
  """Class that montiors a parking garage's available spaces as well as a ticket
  inventory for customers that allows them to pay when they leave the garage

  Atrributes for the class:
    -tickets = expected to be a list
    -outstandingTickets = expected to be a list,
    -usedSpaces = expected to be a list
    -parkingSpaces = expected to be a list
    -currentTicket = expected to be a dictionary, a 'False' value represents
     an unpaid ticket
  """

  def __init__(self, tickets, parkingSpaces, outstandingTickets = [],
               currentTicket = {}, usedSpaces = []):
    self.tickets = tickets
    self.outstandingTickets = outstandingTickets
    self.parkingSpaces = parkingSpaces
    self.currentTicket = currentTicket
    self.usedSpaces = usedSpaces

  def takeTicket(self):
    self.currentTicket[self.tickets[-1]] = False
    self.outstandingTickets.append(self.tickets.pop())
    self.usedSpaces.append(self.parkingSpaces.pop())
    print(f"\nTicket Taken:\n-Available spaces now: {len(self.parkingSpaces)}"
    f"\n-Available tickets now: {len(self.tickets)}")
    

  def payForParking(self):
    parking_duration = int(input("\nHow many hours were you parked in garage for?"
                       "\n(ex. if parked for 2 hours and 30 mins; enter '2')\n"))
    if parking_duration == 0:
      print("\nHave a nice day. Free parking for stays under a hour!")
      self.currentTicket[self.outstandingTickets[0]] = True
    elif parking_duration >=8:
      print(f"\nWe only charge up to 8 hours. Your total is $40.00 .")
      payment = input("Would you like to pay now? (type 'yes'/'no')\n")
      if payment.lower() == 'yes':
        print("Your ticket has been paid, you have 15 minutes to leave the garage.")
        self.currentTicket[self.outstandingTickets[0]] = True
      elif payment.lower() == 'no':
        print("You must pay at gate before leaving.")
      else:
        print("Didn't recognize input, attempt payment again at gate upon exit.")
    elif parking_duration < 8 and parking_duration >=1:
      print(f"Your total is ${parking_duration * 5}.00 .")
      payment_2 = input("Would you like to pay now? (type 'yes'/'no')\n")
      if payment_2.lower() == 'yes':
        print("Your ticket has been paid, you have 15 minutes to leave the garage.")
        self.currentTicket[self.outstandingTickets[0]] = True
      elif payment_2.lower() == 'no':
        print("You must pay at gate before leaving.")
      else:
        print("Didn't recognize input, attempt payment again at gate upon exit.")

      
  def leaveGarage(self):
    if self.currentTicket[self.outstandingTickets[0]] == False:
      payment_exit = input("Ticket is unpaid, must pay before leaving.\nPay now?"
      "('yes'/'no') ")
      if payment_exit.lower() == 'yes':
        print("Your ticket's been paid, thank you and have a good day!")
      elif payment_exit.lower() == 'no':
        print("\nI have no time for crime, you must pay or cops are on their way.")
      else:
        print("ERROR: Please see parking garage attendant.")
    elif self.currentTicket[self.outstandingTickets[0]] == True:
      print("\nTicket has already been paid. Thank you and enjoy the rest of your day!")
    self.parkingSpaces.append(self.usedSpaces.pop())
    self.tickets.append(self.outstandingTickets.pop())
    print(f"Transaction complete:\n-Available spaces now: {len(self.parkingSpaces)}"
    f"\n-Available tickets now: {len(self.parkingSpaces)}")



## Input statment here asks user for the size of their parking garage and then
## forms two lists based off size for the 'tickets' and 'parkingSpaces'
## attributes


garage_size = int(input("Please enter total number of spaces in parking garage. "))
max_tickets = list(range(garage_size))
total_spaces = list(range(garage_size))

atlanta_city_parking = ParkingGarage(max_tickets, total_spaces)

atlanta_city_parking.takeTicket()
atlanta_city_parking.payForParking()
atlanta_city_parking.leaveGarage()


  
```
